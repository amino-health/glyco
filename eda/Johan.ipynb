```python
from datetime import timedelta as tdel
def shift_fwd(t, h=1, m=0):
    return t+ tdel(hours=h, minutes=m)
def shift_back(t, h=1, m=0):
    return t- tdel(hours=h, minutes=m)
```

```python
imgs=list(meals_by_day.loc['2020-10-05'].b)
for x in imgs:
    mymeal = imeals.loc[x]

    mtime = mymeal['a']
    d = mymeal['a'].strftime('%Y-%m-%d')
    mday = booboo.loc[d]
    plt.axvline(x=shift_back(mtime, 1, 30).minute+shift_back(mtime, 1, 30).hour*60, label=x+str(mymeal['hh'])+str(mymeal['a'].minute), color='red')
    

plt.plot(mday.index.minute+mday.index.hour*60, mday[glbl].rolling(5).mean())

plt.xticks(rotation=45)
#plt.ylim(4, 9)

imgs=list(meals_by_day.loc['2020-10-10'].b)
for x in imgs:
    mymeal = imeals.loc[x]

    mtime = mymeal['a']
    d = mymeal['a'].strftime('%Y-%m-%d')
    mday = booboo.loc[d]
    plt.axvline(x=shift_back(mtime, 1, 30).minute+shift_back(mtime, 1, 30).hour*60, label=x+str(mymeal['hh'])+str(mymeal['a'].minute), color='black')
    

plt.plot(mday.index.minute+mday.index.hour*60, mday[glbl].rolling(5).mean())

plt.xticks(rotation=45)
#plt.ylim(4, 9)
plt.show()
```
